Les fonctions de gestion du plateau:

```python
def creer_plateau(taille):
    plateau = []

    for i in range(0, taille):
        ligne = []
        for j in range(0, taille):
            ligne.append(' ')
        plateau.append(ligne)

    return plateau


def affiche_plateau(plateau):
    print("\n")
    for i in range(len(plateau)-1, -1, -1):
        if (i>9):
            print(" ", i, end=' ')
        else: 
            print(" ", i, " ", end='')
        for j in range(0, len(plateau)):
            print("  |", plateau[i][j], end='')
        print("  | ")
    print("        ", 0, end='')
    for k in range(1, len(plateau)):
        if (k<10):
            print("   ", k, end='')
        else:
            print("  ", k, end='')
    print("\n")


def change_joueur(joueurCourant, joueur1, joueur2):
    if (joueurCourant == joueur1):
        joueurCourant = joueur2
    else :
        joueurCourant = joueur1
    return joueurCourant


def initial_pion(plateau, coordonnees):
    coordonnees.insert(random.randint(0, len(plateau)-1), 0)
    coordonnees.insert(random.randint(0, len(plateau)-1), 1)
    while (coordonnees[0] == 0 or coordonnees[1] == 0 or coordonnees[0] == coordonnees[1]):
        coordonnees[0] = random.randint(0, len(plateau)-1)
        coordonnees[1] = random.randint(0, len(plateau)-1)
    plateau[coordonnees[0]][coordonnees[1]] = 'o'


def verif_victoire(plateau):
    fini = False
    if plateau[0][0] == 'o':
            fini = True
    return fini

```

```python
def deplacement_humain(plateau, coordonnees):
    direction = ' '
    nbcases = 0
    ancienp1 = coordonnees[0]
    ancienp2 = coordonnees[1]
    place = False
    
    while (place == False):

        if (coordonnees[1] == 0) :
            direction = 'b'
            print("La seule direction possible est vers le bas.")
        elif (coordonnees[0] == 0) :
            direction = 'g'
            print("La seule direction possible est vers la gauche")
        
        while (direction != 'b' and direction != 'd' and direction != 'g') :
            direction = (input("Direction g(gauche)  b(bas)  d(diagonale) : "))
            
        nbcases = (int)(input("Nombre de case a parcourir:"))
        
        if (direction == 'g') :
            while ((coordonnees[1] - nbcases)<0 or nbcases<=0):
                nbcases = (int)(input("Erreur, veuillez saisir un autre nombre de cases: "))
            coordonnees[1] = (coordonnees[1] - nbcases)
            place = True

        elif (direction == 'b') : 
            while ((coordonnees[0] - nbcases)<0 or nbcases<=0):
                nbcases = (int)(input("Erreur, veuillez saisir un autre nombre de cases: "))
            coordonnees[0] = (coordonnees[0] - nbcases)
            place = True

        elif (direction == 'd') : 
            while ((coordonnees[0] - nbcases)<0 or (coordonnees[1] - nbcases)<0 or nbcases<=0):
                nbcases = (int)(input("Erreur, veuillez saisir un autre nombre de cases: "))
            coordonnees[0] = (coordonnees[0] - nbcases)
            coordonnees[1] = (coordonnees[1] - nbcases)
            place = True
        else:
            print("La direction est incorrect")
    
    #suppression de l'emplacament précédent
    plateau[ancienp1][ancienp2] = ' '
    
    # placement du nouvelle emplacement du pion
    plateau[coordonnees[0]][coordonnees[1]] = 'o'

    return plateau 


def deplacement_aleatoire(plateau, coordonnees):
    direction = ' '
    nbcases = 0
    place = False
    ancienp1 = coordonnees[0]
    ancienp2 = coordonnees[1]    
    
    while (place == False):
        directions = ['g', 'b', 'd']
        d = random.randint(0, 2)
        direction = directions[d]
        
        #l'ordi va vers la direction avec un nombre de cases inférieures 
        #au nombres de cases entre 0 et la position du pion
        if (direction == 'g'):
            if (coordonnees[1] == 1):
                nbcases = 1
            else:
                nbcases = random.randint(1, coordonnees[1])
                while (nbcases == 0):
                    nbcases = random.randint(1, coordonnees[1])

            coordonnees[1] = (coordonnees[1] - nbcases)
            place = True

        elif (direction == 'b'):
            if (coordonnees[0] == 1):
                nbcases = 1
            else:
                nbcases = random.randint(1, coordonnees[0])
                while (nbcases == 0):
                    nbcases = random.randint(1, coordonnees[0])

            coordonnees[0] = (coordonnees[0] - nbcases)
            place = True
            
        elif (direction == 'd'):
            #pour ne pas que l'ordi fasse un nombre de cases trop grande 
            #en diagonale et sorte du plateau
            if (coordonnees[0]<=coordonnees[1]):
                if (coordonnees[0] == 1):
                    nbcases = 1
                else:
                    nbcases = random.randint(1, coordonnees[0])
                    while (nbcases == 0):
                        nbcases = random.randint(1, coordonnees[0])
                coordonnees[0] = (coordonnees[0] - nbcases)
                coordonnees[1] = (coordonnees[1] - nbcases)
                place = True
                
            elif (coordonnees[0]>coordonnees[1]):
                if (coordonnees[1] == 1):
                    nbcases = 1
                else:
                    nbcases = random.randint(1, coordonnees[1])
                    while (nbcases == 0):
                        nbcases = random.randint(1, coordonnees[1])
                coordonnees[0] = (coordonnees[0] - nbcases)
                coordonnees[1] = (coordonnees[1] - nbcases)
                place = True
                
    #suppression de l'emplacament précédent
    plateau[ancienp1][ancienp2] = ' '
    
    # placement du nouvelle emplacement du pion
    plateau[coordonnees[0]][coordonnees[1]] = 'o'

    return plateau 


def deplacement_ordi(plateau, coordonnees):
    position = position_gagnante(plateau)
    limite = False
    change = False
    i = 0

    ancienp1 = coordonnees[0]
    ancienp2 = coordonnees[1]

    
    while (i<len(position) and limite == False and change == False):
        
        #si le place gagnante est trop grande par rapport à la place du pion 
		#alors on arrête les test
        if (coordonnees[0] >= position[i][0] and coordonnees[1] >= position[i][0] or coordonnees[0] >= position[i][1] and coordonnees[1] >= position[i][1]):
            
            #si le pion est déjà sur une position gagnante
			#alors faire un déplacement aléatoire
            if (coordonnees[0] == position[i][0] and coordonnees[1] == position[i][0] or coordonnees[0] == position[i][1] and coordonnees[1] == position[i][1]):
                change = False
                
            #si le pion est sur la même abscisse qu'une position gagnante de la moitié supérieur
            #et que l'ordonnée du pion est plus grand que l'ordonnée de la position gagnante
            elif (coordonnees[0] == position[i][0] and coordonnees[1]>position[i][1] and change == False):
                coordonnees[1] = position[i][1]
                change = True
                
            #si le pion est sur la même abscisse qu'une position gagnante de la moitié inférieur
            #et l'ordonnée du pion est plus grand que l'ordonnée de la position gagnante
            elif (coordonnees[0] == position[i][1] and coordonnees[1]>position[i][0] and change == False):
                coordonnees[1] = position[i][0]
                change = True
                
            #si le pion est sur le même ordonnée d'une position gagnante de la moitié inférieur
            #et l'abscisse du pion est plus grand que l'abscisse de la position gagnante 
            elif (coordonnees[1] == position[i][0] and coordonnees[0]>position[i][1] and change == False):
                coordonnees[0] = position[i][1]
                change = True
                
            #si le pion est sur le même ordonnée qu'une position gagnante de la moitié supérieur
            #et l'abscisse du pion est plus grand que l'abscisse de la position gagnante 
            elif (coordonnees[1] == position[i][1] and coordonnees[0]>position[i][0] and change == False):
                coordonnees[0] = position[i][0]
                change = True
                
            #si le pion est sur la diagonale pour aller à (0,0)
            elif (coordonnees[0] == coordonnees[1] and change == False):
                coordonnees[0] = 0
                coordonnees[1] = 0
                change = True
                
            #si une des positions du pion est sur la ligne/colonne 0
            #autrement dit peut aller directement à (0,0)
            elif (coordonnees[0]==0 or coordonnees[1]==0 and change == False):
                coordonnees[0] = 0
                coordonnees[1] = 0
                change = True
            
            #si en faisant une diagonale le pion arrive sur une position gagnante
            p1 = coordonnees[0]-position[i][0]
            p2 = coordonnees[1]-position[i][1]
            if ((p1 == p2) and (p1!= 0) and change == False):
                coordonnees[0] = position[i][0]
                coordonnees[1] = position[i][1]
                change = True

            #si en faisant une diagonale le pion arrive sur une position gagnante
            p3 = coordonnees[0]-position[i][1]
            p4 = coordonnees[1]-position[i][0]
            if ((p3 == p4) and (p3!= 0) and change == False):
                coordonnees[0] = position[i][1]
                coordonnees[1] = position[i][0]
                change = True
                
        else:
            limite = True

        i+=1
    if (change == False):
        deplacement_aleatoire(plateau, coordonnees)
    
    #suppression de l'emplacament précédent
    plateau[ancienp1][ancienp2] = ' '
    
    # placement du nouvelle emplacement du pion
    plateau[coordonnees[0]][coordonnees[1]] = 'o'

    dep_ordi([coordonnees[0], coordonnees[1], prob])


def position_gagnante(plateau):
    position = [[0, 0]]
    limite = False
    i = 1
    nb = 1
    
    while (i < len(plateau) and limite == False):
        if (nb+i < len(plateau)):
            position.append([nb, nb+i])
            j = 0
            while (j <= i):
                if (nb == position[j][0]):
                    nb+=1
                    j=0
                if (nb == position[j][1]):
                    nb+=1
                    j=0
                j+=1
        else:
            limite = True
        i+=1
    return position
    
```
